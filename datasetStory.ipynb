```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('adult.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
sns.set(color_codes=True)
x,y = 'race', 'income'
g = (df.groupby(x)[y].value_counts(normalize=True).mul(100).rename('percent').reset_index()
     .pipe((sns.catplot,'data'), x=x,y='percent',hue=y,kind='bar'))
g.set_xticklabels(rotation=90)
```
